## 設置

執行以下設置 cell 以載入你的 API 金鑰並建立 `get_completion` 輔助函數。


```python
%pip install anthropic

# Import python's built-in regular expression library
import re
import anthropic

# Retrieve the API_KEY & MODEL_NAME variables from the IPython store
%store -r API_KEY
%store -r MODEL_NAME

client = anthropic.Anthropic(api_key=API_KEY)

# Note that we changed max_tokens to 4K just for this lesson to allow for longer completions in the exercises
def get_completion(prompt: str, system_prompt=""):
    message = client.messages.create(
        model=MODEL_NAME,
        max_tokens=4000,
        temperature=0.0,
        system=system_prompt,
        messages=[
          {"role": "user", "content": prompt}
        ]
    )
    return message.content[0].text
```

## 課程

**Claude 對清楚直接的指示回應最好。**

把 Claude 想像成任何一個新入職的人類。**Claude 沒有任何情境**，除了你明確告訴它的之外。就像你第一次指導人類完成任務時一樣，你解釋得越多，Claude 的回應就越好。

當需要沒有歧義的明確回應時，只需直接要求 Claude 提供！不要讓 Claude 自行判斷是否需要固執或確定——只需告訴它。


### 範例

讓我們以寫詩為例。（忽略任何音節不匹配——LLM 在計算音節方面還不太好。）


```python
# Prompt
PROMPT = "Write a haiku about robots."

# Print Claude's response
print(get_completion(PROMPT))
```

這首俳句已經很好了，但用戶可能希望 Claude 直接進入詩歌，不要有「Here is a haiku」的前言。

我們如何實現這一點？我們**直接要求**！


```python
# Prompt
PROMPT = "Write a haiku about robots. Skip the preamble; go straight into the poem."

# Print Claude's response
print(get_completion(PROMPT))
```

這是另一個範例。讓我們問 Claude 誰是有史以來最好的籃球運動員。你可以在下面看到，雖然 Claude 列出了幾個名字，**但它沒有以明確的「最佳」回應**。


```python
# Prompt
PROMPT = "Who is the best basketball player of all time?"

# Print Claude's response
print(get_completion(PROMPT))
```

我們能讓 Claude 下定決心並決定最佳球員嗎？可以！只需要求！


```python
# Prompt
PROMPT = "Who is the best basketball player of all time? Yes, there are differing opinions, but if you absolutely had to pick one player, who would it be?"

# Print Claude's response
print(get_completion(PROMPT))
```

### 練習 2.1 - 西班牙語
修改 `SYSTEM_PROMPT` 讓 Claude 以西班牙語輸出答案。


```python
# System prompt - this is the only field you should change
SYSTEM_PROMPT = "[Replace this text]"

# Prompt
PROMPT = "Hello Claude, how are you?"

# Get Claude's response
response = get_completion(PROMPT, SYSTEM_PROMPT)

# Function to grade exercise correctness
def grade_exercise(text):
    return "hola" in text.lower()

# Print Claude's response and the corresponding grade
print(response)
print("\n--------------------------- GRADING ---------------------------")
print("This exercise has been correctly solved:", grade_exercise(response))
```

### 練習 2.2 - 只選一個球員

修改 `PROMPT` 讓 Claude 完全不猶豫，只回應**一個具體球員的名字**，**沒有其他文字或標點符號**。


```python
# Prompt - this is the only field you should change
PROMPT = "[Replace this text]"

# Get Claude's response
response = get_completion(PROMPT)

# Function to grade exercise correctness
def grade_exercise(text):
    return text == "Michael Jordan"

# Print Claude's response and the corresponding grade
print(response)
print("\n--------------------------- GRADING ---------------------------")
print("This exercise has been correctly solved:", grade_exercise(response))
```

### 練習 2.3 - 寫一個故事

修改 `PROMPT` 讓 Claude 回應盡可能長的內容。如果你的答案**超過 800 個字**，Claude 的回應將被評為正確。


```python
# Prompt - this is the only field you should change
PROMPT = "[Replace this text]"

# Get Claude's response
response = get_completion(PROMPT)

# Function to grade exercise correctness
def grade_exercise(text):
    trimmed = text.strip()
    words = len(trimmed.split())
    return words >= 800

# Print Claude's response and the corresponding grade
print(response)
print("\n--------------------------- GRADING ---------------------------")
print("This exercise has been correctly solved:", grade_exercise(response))
```
