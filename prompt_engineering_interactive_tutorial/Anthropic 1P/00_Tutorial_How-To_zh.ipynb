## 使用說明與技巧 💡

- 本課程使用 temperature 0 的 Claude 3 Haiku。我們會在課程後面詳細介紹 temperature。現在，只需了解這些設定會產生更具決定性的結果。
- 在整個課程中，你需要解決練習題。解題方法很多，所以只要通過練習題的評估，你的答案就是正確的。
- 每個練習題旁邊都有提示按鈕。


```python
import anthropic

client = anthropic.Anthropic(api_key=API_KEY)

def get_completion(prompt: str):
    message = client.messages.create(
        model=MODEL_NAME,
        max_tokens=2000,
        temperature=0.0,
        messages=[
          {"role": "user", "content": prompt}
        ]
    )
    return message.content[0].text
```

現在我們將為 Claude 撰寫一個範例 prompt，並透過執行 `get_completion` 輔助函數來印出 Claude 的輸出。執行以下 cell 將在其下方印出 Claude 的回應。

請隨意修改 prompt 字串，以引出 Claude 不同的回應。


```python
# Prompt
prompt = "Hello, Claude!"

# Get Claude's response
print(get_completion(prompt))
```

前面定義的 `API_KEY` 和 `MODEL_NAME` 變數將在整個教學中使用。確保從頂部到底部執行每個教學頁面的 cell。
