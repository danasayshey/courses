## 設置

執行以下設置 cell 以載入你的 API 金鑰並建立 `get_completion` 輔助函數。


```python
%pip install anthropic

# Import python's built-in regular expression library
import re
import anthropic

# Retrieve the API_KEY & MODEL_NAME variables from the IPython store
%store -r API_KEY
%store -r MODEL_NAME

client = anthropic.Anthropic(api_key=API_KEY)

def get_completion(prompt: str, system_prompt=""):
    message = client.messages.create(
        model=MODEL_NAME,
        max_tokens=2000,
        temperature=0.0,
        system=system_prompt,
        messages=[
          {"role": "user", "content": prompt}
        ]
    )
    return message.content[0].text
```

## 課程

Anthropic 提供兩個 API：舊版的 [Text Completions API](https://docs.anthropic.com/claude/reference/complete_post) 和目前的 [Messages API](https://docs.anthropic.com/claude/reference/messages_post)。在本教學中，我們將專門使用 Messages API。

最少，Messages API 呼叫需要以下參數：
- `model`：你打算呼叫的模型 API 名稱
- `max_tokens`：在停止生成之前要生成的最大 token 數量。請注意，Claude 可能在達到最大值之前就停止。此參數只設置絕對的最大 token 數量。
- `messages`：一個輸入訊息的陣列。

`messages` 參數是一個訊息物件的陣列，每個物件都有 `role` 和 `content`。訊息必須在 `user` 和 `assistant` 角色之間交替，並且必須以 `user` 角色開始。


### 範例

讓我們看看 Claude 如何回應一些格式正確的 prompt。對於以下每個 cell，執行 cell（`shift+enter`），Claude 的回應將顯示在 block 下方。


```python
# Prompt
PROMPT = "Hi Claude, how are you?"

# Print Claude's response
print(get_completion(PROMPT))
```

```python
# Prompt
PROMPT = "Can you tell me the color of the ocean?"

# Print Claude's response
print(get_completion(PROMPT))
```

```python
# Prompt
PROMPT = "What year was Celine Dion born in?"

# Print Claude's response
print(get_completion(PROMPT))
```

現在讓我們看看一些不包含正確 Messages API 格式的 prompt。對於這些格式錯誤的 prompt，Messages API 會返回錯誤。

首先，我們有一個 Messages API 呼叫的範例，其 `messages` 陣列中缺少 `role` 和 `content` 欄位。


```python
# Get Claude's response
response = client.messages.create(
        model=MODEL_NAME,
        max_tokens=2000,
        temperature=0.0,
        messages=[
          {"Hi Claude, how are you?"}
        ]
    )

# Print Claude's response
print(response[0].text)
```

以下是一個未能在 `user` 和 `assistant` 角色之間交替的 prompt。


```python
# Get Claude's response
response = client.messages.create(
        model=MODEL_NAME,
        max_tokens=2000,
        temperature=0.0,
        messages=[
          {"role": "user", "content": "What year was Celine Dion born in?"},
          {"role": "user", "content": "Also, can you tell me some other facts about her?"}
        ]
    )

# Print Claude's response
print(response[0].text)
```

`user` 和 `assistant` 訊息**必須交替**，且訊息**必須以 `user` turn 開始**。你可以在 prompt 中有多個 `user` 和 `assistant` 配對（如模擬多輪對話）。你也可以在最後的 `assistant` 訊息中填入文字，讓 Claude 從你留下的地方繼續。


#### System Prompt 範例


```python
# System prompt
SYSTEM_PROMPT = "Your answer should always be a series of critical thinking questions that further the conversation (do not provide answers to your questions). Do not actually answer the user question."

# Prompt
PROMPT = "Why is the sky blue?"

# Print Claude's response
print(get_completion(PROMPT, SYSTEM_PROMPT))
```

為什麼要使用 system prompt？**寫得好的 system prompt 可以多方面提升 Claude 的效能**，例如提高 Claude 遵守規則和指示的能力。

### 練習 1.1 - 數到三
使用正確的 `user` / `assistant` 格式，編輯下面的 `PROMPT` 讓 Claude **數到三**。輸出也會顯示你的解答是否正確。


```python
# Prompt - this is the only field you should change
PROMPT = "[Replace this text]"

# Get Claude's response
response = get_completion(PROMPT)

# Function to grade exercise correctness
def grade_exercise(text):
    pattern = re.compile(r'^(?=.*1)(?=.*2)(?=.*3).*$', re.DOTALL)
    return bool(pattern.match(text))

# Print Claude's response and the corresponding grade
print(response)
print("\n--------------------------- GRADING ---------------------------")
print("This exercise has been correctly solved:", grade_exercise(response))
```

### 練習 1.2 - System Prompt

修改 `SYSTEM_PROMPT` 讓 Claude 表現得像一個 3 歲的小孩。


```python
# System prompt - this is the only field you should change
SYSTEM_PROMPT = "[Replace this text]"

# Prompt
PROMPT = "How big is the sky?"

# Get Claude's response
response = get_completion(PROMPT, SYSTEM_PROMPT)

# Function to grade exercise correctness
def grade_exercise(text):
    return bool(re.search(r"giggles", text) or re.search(r"soo", text))

# Print Claude's response and the corresponding grade
print(response)
print("\n--------------------------- GRADING ---------------------------")
print("This exercise has been correctly solved:", grade_exercise(response))
```
